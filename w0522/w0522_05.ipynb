```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # windows 한글폰트 연결
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac 한글폰트 연결
matplotlib.rcParams['font.size'] = 15                 # 폰트 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False     # 그래프 마이너스 표시
```

```python
### 꺽은선 그래프
# 1. 제목 : 월별매출액
# 2. x축 문구 - 월별 색상 green
# 3. y축 문구 - 매출액 red
# 4. x축 간격 1,1.5,2,2.5,3,3.5,4,4.5,5
# 5. y축 간격 10,20,30

x = [1,2,3,4,5]
y = [10,20,22,25,29]

plt.plot(x,y)
plt.title("월별매출액")
plt.xlabel('월별',color='g',loc='right')
plt.ylabel('매출액',color='r',loc='bottom')
plt.xticks([1,1.5,2,2.5,3,3.5,4,4.5,5])
plt.yticks([10,20,30])
plt.show()

```
