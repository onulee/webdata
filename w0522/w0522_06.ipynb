```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # windows 한글폰트 연결
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac 한글폰트 연결
matplotlib.rcParams['font.size'] = 15                 # 폰트 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False     # 그래프 마이너스 표시
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
df = pd.read_excel('data/score.xlsx',index_col='지원번호')
```

```python
# label:범례, linewidth : 선의 두께
# marker : 지점에 마커표시 - o,v,x,s,
# markeredgecolor: 마커테두리 mec, markerfacecolor mfc : 마커색상
# 선형태 - linestyle='None': 선제거, : 점선, -- 대쉬선, -. 점선점
plt.plot(x,y,label='서울',linewidth=2,marker='o',markersize=10,
         mec='red',mfc='yellow', linestyle='-.'
         )
# plt.plot(x,y,label='서울',linewidth=2,marker='o',markersize=10,
#          markeredgecolor='red',markerfacecolor='yellow'
#          )
# plt.plot(x,y,label='서울',linewidth=2,marker='o',markersize=10,linestyle='None')  # 꺽은선 그래프
# plt.plot(df['이름'],df['국어'])  # 리스트 입력시 그래프 나타남.
plt.title('그래프')  # 한글지원부분 설정
# loc 범례위치 - y축 : upper,center,lower, x축: left,center,right
# 위치점 지정 loc=(0.5,0.5) # 가운데지점
# plt.legend(loc=(0.25,0.5))  
plt.legend()
plt.show() 
```

```python
plt.plot(df['이름'],df['국어'],label='국어',marker='o',linestyle='None')
plt.plot(df['이름'],df['영어'],label='영어',marker='v',linestyle='None')
plt.plot(df['이름'],df['수학'],label='수학',marker='X')
plt.title("학생성적그래프")
plt.legend()
plt.show()
```
