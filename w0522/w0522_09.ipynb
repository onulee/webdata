```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # windows 한글폰트 연결
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac 한글폰트 연결
matplotlib.rcParams['font.size'] = 15                 # 폰트 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False     # 그래프 마이너스 표시
```

```python
day = [1,2,3]
kor = [100,80,90]
eng = [95,85,92]
math = [80,90,70]
```

여러개 그래프 그리기

```python
plt.plot(day,kor,label='국어',marker='o',color='#063970')
plt.plot(day,eng,label='영어',marker='v',linestyle='--')
plt.plot(day,math,label='수학',marker='X',linestyle=':')
# plt.bar(day,math,label='수학')
plt.legend(loc='upper center',ncol=3)   # ncol = 3개를 1행으로 보여줌
for i,v in enumerate(kor):
    plt.text(day[i],kor[i]+1,v,ha='center')
    plt.text(day[i]+0.1,eng[i],eng[i],ha='center')
    plt.text(day[i],math[i]+1,math[i],ha='center')
plt.yticks([60,70,80,90,100,110])    
plt.xticks([1,2,3,3.2])    
plt.show()
```

```python
name = ['홍길동','유관순','이순신']
kor = [100,80,90]
eng = [95,85,92]
math = [80,90,70]
```

막대그래프

```python
# plt.plot(name,kor)
colors = ['r','g','b']
# color : 막대그래프 색상 - 배열로 입력하면 그래프 각각의 색상을 넣을수 있음.
# alpha : 투명도 지정
plt.bar(name,kor,label='국어', color=colors, alpha=0.5)
plt.plot(name,eng,marker='o')
plt.yticks([70,80,90,100])  # 눈금표시
plt.ylim(70,105)   # 눈금의 범위 지정
plt.legend()
for i,v in enumerate(kor):
    plt.text(name[i],kor[i]-5,kor[i],ha='center')
plt.show()
```

```python
colors = ['r','g','b']
plt.bar(name,kor,label='국어', color=colors, alpha=0.7)
plt.xticks(rotation=90)
plt.yticks([70,80,90,100])  
plt.ylim(70,105)   
plt.legend()

plt.show()
```
