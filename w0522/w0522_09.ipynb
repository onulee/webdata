```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # windows 한글폰트 연결
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac 한글폰트 연결
matplotlib.rcParams['font.size'] = 15                 # 폰트 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False     # 그래프 마이너스 표시
```

```python
day = [1,2,3]
kor = [100,80,90]
eng = [95,85,92]
math = [80,90,70]
```

여러개 그래프 그리기

```python
plt.plot(day,kor,label='국어',marker='o',color='#063970')
plt.plot(day,eng,label='영어',marker='v',linestyle='--')
plt.plot(day,math,label='수학',marker='X',linestyle=':')
# plt.bar(day,math,label='수학')
plt.legend(loc='upper center',ncol=3)   # ncol = 3개를 1행으로 보여줌
for i,v in enumerate(kor):
    plt.text(day[i],kor[i]+1,v,ha='center')
    plt.text(day[i]+0.1,eng[i],eng[i],ha='center')
    plt.text(day[i],math[i]+1,math[i],ha='center')
plt.yticks([60,70,80,90,100,110])    
plt.xticks([1,2,3,3.2])    
plt.show()
```

```python
name = ['홍길동','유관순','이순신']
# no = [1,2,3] # 리스트 타입
no = np.arange(1,4) #numpy 배열형태로 적용
kor = [100,80,90]
eng = [95,85,92]
math = [80,90,77]
```

```python
plt.bar(no,kor,width=0.2)
for i,v in enumerate(kor):
    plt.text(no[i],kor[i],kor[i],ha='center')
```

```python
colors = ['r','g','b'] 
### width : 바의 사이즈 조절
### no numpy배열 형태
w=0.2
# bar(x좌표,y좌표)
bar = plt.bar(no-w,kor,width=0.2)
### set_hatch : bar 무늬 적용
bar[0].set_hatch('/')  # bar 무늬 적용
bar[1].set_hatch('x')  
bar[2].set_hatch('..')
plt.bar(no,eng,width=0.2)
plt.bar(no+w,math,width=0.2)
### 각각의 막대그래프 위에 점수를 출력하시오.
for i,v in enumerate(kor):
    plt.text(no[i]-w,kor[i]+1,kor[i],ha='center')
    plt.text(no[i],eng[i]+1,eng[i],ha='center')
    plt.text(no[i]+w,math[i]+1,math[i],ha='center')
plt.ylim(70,105)
plt.xticks(no,name)  # x좌표 이동에 필요한 수치값을 no를 사용할수 있음.

 
plt.show()
```

```python
name = ['홍길동','유관순','이순신']
# no = [1,2,3] # 리스트 타입
no = np.arange(1,4) #numpy 배열형태로 적용
kor = [100,80,90]
eng = [95,85,92]
math = [80,90,77]
```

```python
# 3개의 막대 그래프를 출력하시오. grid : -- 색상은 검정 투명도0.1 
# 3개 막대 그래프 위에 점수를 출력하시오.

colors = ['r','g','b'] 
### width : 바의 사이즈 조절
w=0.2
# bar(x좌표,y좌표)
bar = plt.bar(no-w,kor,width=0.2)
plt.bar(no,eng,width=0.2)
plt.bar(no+w,math,width=0.2)
for i,v in enumerate(kor):
    plt.text(no[i]-w,kor[i]+1,kor[i],ha='center')
    plt.text(no[i],eng[i]+1,eng[i],ha='center')
    plt.text(no[i]+w,math[i]+1,math[i],ha='center')
plt.ylim(70,105)
plt.xticks(no,name)  # x좌표 이동에 필요한 수치값을 no를 사용할수 있음.
plt.show()
```
