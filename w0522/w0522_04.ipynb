```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # windows 한글폰트 연결
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac 한글폰트 연결
matplotlib.rcParams['font.size'] = 15                 # 폰트 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False     # 그래프 마이너스 표시
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)  # plot(x축데이터, y축데이터)
plt.title('온도조절 그래프')  # 상단 제목 입력
# x축 설명문구, color : 색상지정:색상값,16진수코드,약식색상, loc = left,center,right
plt.xlabel('X축 설명 - 월별',color='red',loc='right' )  
# y축 설명문구 loc=top,center,bottom
plt.ylabel('Y축 설명 - 온도',color='#00aa00',loc="top")  
# plt.ylabel('Y축 설명 - 온도',color='indigo',loc="top")  
# plt.ylabel('Y축 설명 - 온도',color='g')  # y축 설명문구
# plt.xlabel('X축 설명 - 월별')  # x축 설명문구
# plt.ylabel('Y축 설명 - 온도')  # y축 설명문구
plt.xticks([1,2,3])  # x축표시 간격
plt.yticks([2,4,8,10,12,16])  # y축표시 간격
# plt.yticks([2,4,6,8,9,10,11])  # y축표시 간격
# plt.xticks([1,1.2,1.5,2,3])  # x축표시 간격


plt.show()
```
