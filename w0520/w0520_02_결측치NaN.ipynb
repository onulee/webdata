```python
import pandas as pd
import numpy as np    # 데이터처리 라이브러리
```

```python
df = pd.read_csv('data/stuscore.csv',index_col='번호')
df

## 이름 a,A가 들어가면서 영어점수가 80점 이상인 학생을 검색하시오.
filt = (df['이름'].str.lower().str.contains("a")) & (df['영어']>=80)
df[filt]

##  홍길동,홍길순,유관순 학생을 검색하시오.
arr = ['홍길동','홍길순','유관순']
filt = df['이름'].isin(arr)
df[filt]
```

```python
### SW특기에서 p,P가 포함되어 있는 문자를 모두 출력하시오.
# score.xlsx 에서 검색해야 함.
df = pd.read_excel('data/score.xlsx',index_col='지원번호')
filt = df['SW특기'].str.lower().str.contains('p',na=False)
df[filt]
```

```python
df = pd.read_excel('data/score.xlsx',index_col='지원번호')
df['학교'] = np.nan   # 학교컬럼 모두 NaN처리
```

```python
### 결측치 fillna,dropna - axis:index,columns,how:any,all)
# fillna : 모든 NaN 데이터 처리  
# df.fillna('')  # 모든 NaN데이터 처리
# df.fillna("없음",inplace=True)  # inplace처리를 해야 반영됨.

# dropna : 모든 NaN데이터 삭제, NaN데이터가 있는 row행을 삭제
# inplace를 해야 반영됨.
# df.dropna(inplace=True)   # 모두 삭제
# df.dropna(axis='index',how='any')  # index방향,1개만 NaN이 있어도 삭제
# df.dropna(axis='index',how='all')  # index방향,모두 NaN이 있어야 삭제
# df.dropna(axis='columns',how='any')  # columns방향, 1개만 NaN이 있어도 삭제
# df.dropna(axis='columns',how='all')    # columns방향, 모두 NaN이어야 삭제
```

```python
### 국어점수를 모두 NaN 처리후 50으로 모두 변경하시오.
# df['국어'] = np.nan
# df['국어'].fillna(50,inplace=True)

# 1.국어 평균점수를 먼저 구하고, 
# 2.국어 점수 모두 NaN처리, 
# 3.평균점수로 변경
# 국어점수를 NaN처리후 이전국어점수의 평균점수로 변경하시오.
kor = df['국어'].mean()
kor
df['국어'] = np.nan
df['국어'].fillna(kor,inplace=True)
```
