```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('data/score.xlsx',index_col='지원번호')
```

```python
df['합계'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df['평균'] = df['합계'] / 5
df['등급'] = '불합격'
```

```python
filt = df['평균']>=90
df.loc[filt,'등급'] = '합격'
df
```

```python
# int타입에 cm문자 추가방법
# df['키'] = df['키'].astype(str) + 'cm'

# 함수 사용해서 추가
# def add_cm(height):
#     return str(height)+'cm'
# df['키'] = df['키'].apply(add_cm)



# 함수를 사용해서
# 90점 이상 합격, 90점 미만 불합격
# df.loc[df['평균']>=90,'등급'] = '합격'

### 
def rank(score):
    result = ""
    if score >=90: result = "합격"
    else: result = "불합격"
    return result 

df['등급'] = df['평균'].apply(rank)
```

```python
### 90점이상 A, 80점이상 B, 70점이상 C, 60점이상 D,60점미만 F
def rank(score):
    result = ""
    if score >=90: result = "A"
    elif score >=80: result = "B"
    elif score >=70: result = "C"
    elif score >=60: result = "D"
    else: result = "F"
    return result 

df['등급'] = df['평균'].apply(rank)
df
```

```python
### 함수를 사용해서 요일 0:일, 1:월, 2:화, 3:수, 4:목, 5:금, 6:토
### 변경하시오.
def change_week(wData):
    ch = ""
    if wData == 0: ch='일요일' 
    elif wData == 1: ch='월요일'
    elif wData == 2: ch='화요일'
    elif wData == 3: ch='수요일'
    elif wData == 4: ch='목요일'
    elif wData == 5: ch='금요일'
    else: ch='토요일'
    return ch

df['요일'] = df['요일'].apply(change_week)
df
```
