```python
import pandas as pd
df = pd.read_csv('data/stuscore.csv',index_col='번호')
```

str함수

```python
### startswith 함수 - 해당글자로 시작되는 문자 검색
filt = df['이름'].str.startswith('S')
filt = df['이름'].str.startswith('홍')
df[filt]
```

```python
### endswith - 해당글자로 끝나는 문자 검색
df[df['이름'].str.endswith('순')]
df[df['이름'].str.endswith('n')]
```

```python
# contains - 해당글자가 포함되어 있는 문자 검색
df[df['이름'].str.contains('s')]
df[df['이름'].str.contains('순')]
df[~df['이름'].str.contains('순')]   # ~ not을 의미
```

```python
### a가 들어가 있는 문자를 검색하시오. - 40
df[df['이름'].str.contains('a')].count()  # 40개
df[df['이름'].str.contains('A')]          # 8개
df[df['이름'].str.contains('A') | df['이름'].str.contains('a')].count() # 48개
```

소문자로 변경후 글자가 있는지 검색

```python
df['이름'].str.lower()  # 모든 글자 소문자로 변경
# 모든 글자를 소문자로 변경후 a가 포함되어 있는 문자 검색
df[df['이름'].str.lower().str.contains('a')].count()

## 모든 글자 대문자로 변경
df['이름'].str.upper()
df[df['이름'].str.upper().str.contains('A')].count()

```
